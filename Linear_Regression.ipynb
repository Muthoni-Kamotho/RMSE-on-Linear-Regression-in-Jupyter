```python
predict=(lambda date: linear_regression.coef_ * date + linear_regression.intercept_)
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
mse= mean_squared_error( expected, predicted)
rmse = math.sqrt(mse)
```

```python
#default RMSE
rmse
```

```python
#RMSE using different ratios and random state 11
ratios = [(50, 50), (75, 25), (80, 20), (90, 10)]
for ratio in ratios:
    X_train, X_test, y_train, y_test = train_test_split(nyc.Date.values.reshape(-1,1), nyc.Temperature.values, test_size=ratio[1]/100, random_state=11)

    # Training of the linear regression model
    linear_regression = LinearRegression()
    linear_regression.fit(X_train, y_train)
    # Calculate the predicted and expected values
    predicted = linear_regression.predict(X_test)
    expected = y_test
    # Calculate the RMSE
    mse= mean_squared_error( y_test, predicted)
    rmse = math.sqrt(mse)
    rmse
    print(f"train-test_split ratio: {ratio[0]}:{ratio[1]}")
    print(f"THE RMSE: {rmse:}")
    print()
```

The RMSE is used to measure the difference between predicted values and actual values in a regression problem.
